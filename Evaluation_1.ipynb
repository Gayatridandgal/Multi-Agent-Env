```python
!pip install -U --quiet google-genai

from google import genai
import time


API_KEY = ""
client = genai.Client(api_key=API_KEY)


def model_reply(prompt, model_name="gemini-2.5-flash"):
    """Send a prompt to Gemini 2.5 Flash and return text output."""
    response = client.models.generate_content(
        model=model_name,
        contents=prompt
    )
    return response.text.strip()

class Agent:
    def __init__(self, name, role):
        self.name = name
        self.role = role

    def reply(self, message):
        if self.role == "summarizer":
            prompt = (
                f"You are {self.name}, a helpful summarizer. "
                f"shorten this text into 1 concise sentences. keep it short and effective:\n\n"
                f"{message}"
            )
        elif self.role == "negotiator":
            prompt = (
                f"You are {self.name}, a friendly negotiator helping plan a meeting. "
                f"Reply naturally, suggesting or confirming a time :\n\n"
                f"{message}"
            )
        else:
            prompt = f"{self.name} responds briefly to: {message}"

        reply = model_reply(prompt)
        return f"{self.name}: {reply}"

```

```python

print("Scenario 1: Two Agents Collaborate to Summarize\n")

agent1 = Agent("Agent A", "summarizer")
agent2 = Agent("Agent B", "summarizer")

paragraph = (
    "Artificial Intelligence enables machines to perform tasks that typically "
    "require human intelligence, such as reasoning, learning, and perception."
)

print("Original Text:\n", paragraph, "\n")

message = paragraph
for i in range(2):
    speaker = agent1 if i % 2 == 0 else agent2
    message = speaker.reply(message)
    print(message)
    time.sleep(1)

print("\n✅ Summary complete!")
print("=" * 70)

```

```python


print("\n Scenario 2: Two Agents Negotiate a Meeting Time\n")

agent3 = Agent("Agent A", "negotiator")
agent4 = Agent("Agent B", "negotiator")

message = "Let's schedule our team meeting tomorrow at 9 PM."
print("Initial Proposal:\n", message, "\n")

for i in range(3):
    speaker = agent3 if i % 2 == 0 else agent4
    message = speaker.reply(message)
    print(message)
    time.sleep(1)

print("\n✅ Negotiation completed!")
```
